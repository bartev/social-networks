```python
%matplotlib widget
import matplotlib.pyplot as plt

import networkx as nx
import numpy as np
import pandas as pd
```

```python
G1 = nx.Graph()
G1.add_edges_from(
    [
        (0, 1),
        (0, 2),
        (0, 3),
        (0, 5),
        (1, 3),
        (1, 6),
        (3, 4),
        (4, 5),
        (4, 7),
        (5, 8),
        (8, 9),
    ]
)
nx.draw_networkx(G1)
```

# Adjacency list

```python
for line in nx.generate_adjlist(G1):
    print(line)
```

```python
with open("G1_adjlist.txt", "w") as f:
    for line in nx.generate_adjlist(G1):
        f.write(line + "\n")
```

The 1st line is the source node

The 2nd line is target nodes

# Adjacency Matrix

```python
# Create a 10x10 array with random 0s and 1s
array = np.random.randint(0, 2, size=(5,5))

G_mat = np.array(array)

```

```python
G_mat
```

Convert an adjacency matrix into a network graph

```python
G3 = nx.Graph(G_mat)
sorted(G3.edges())
```

# Edgelist

```python
# Save sorted edges to a file
with open("edges.txt", "w") as f:
    for edge in sorted(G3.edges()):
        f.write(f"{edge[0]} {edge[1]}\n")
```
